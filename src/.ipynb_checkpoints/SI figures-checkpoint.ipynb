## Temporal distribution

```python
df = pd.read_pickle('../dataset/03_Final_dataframe/df_ddo_including_only_truebeliefs_nodup(N192307).p')
```

```python
date_counts = df['debate_date'].value_counts().sort_index()

weekly_counts = df.resample('W', on='debate_date').size()
monthly_counts = df.resample('M', on='debate_date').size()
```

```python
df_unique = df.drop_duplicates(subset='debate_key')
debate_weekly_counts = df_unique.resample('W', on='debate_date').size()
```

```python
len(df_unique)
```

```python
def get_color_from_palette(ratio, cmap_name='viridis'):
    cmap = plt.get_cmap(cmap_name)
    color = cmap(ratio)  # Get RGBA color
    return matplotlib.colors.rgb2hex(color)
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10,6))

ax = axes[0]
ax.plot(debate_weekly_counts.index, debate_weekly_counts.values, '-', color=get_color_from_palette(0.3))
ax.set_title('Weekly number of debates')
ax.set_xlabel('Date', fontsize=15)
ax.set_ylabel('Debates', fontsize=15)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability


ax = axes[1]
ax.plot(weekly_counts.index, weekly_counts.values, '-', color=get_color_from_palette(0.55))
ax.set_title('Weekly user participation over time (debaters + voters)')
ax.set_xlabel('Date', fontsize=15)
ax.set_ylabel('Participations', fontsize=15)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

plt.tight_layout()  # Adjust layout to make room for label rotation
plt.savefig('../fig/SI/Temporal_distribution.pdf', transparent=True, bbox_inches='tight')
plt.show()
```

## User and Debate statistics

```python
def logbinning(data, numbin=40):
    # log-scaled bins
    bins = np.logspace(0, np.log2(max(data)), num=numbin, base=2)
    widths = (bins[1:] - bins[:-1])

    # Calculate histogram
    hist = np.histogram(data, bins=bins)
    # normalize by bin width
    hist_norm = hist[0]/widths

    # plot it!
    x = bins[:-1]
    y = hist_norm
    
    return x, y, widths
```

```python
num_debates = []
for g, dat in df.groupby('username'):
    num_debates.append(len(dat))
```

```python
num_users = []
for g, dat in df.groupby('debate_title'):
    num_users.append(len(dat))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))

#ax1
ax = axes[0]
x, y, widths = logbinning(num_debates)
ax.loglog(x,y,'o')
ax.set_xlabel('Number of debates per user')
ax.set_ylabel('Normalized count') #number of data points per unit width

#ax2
ax = axes[1]
x, y, widths = logbinning(num_users)
ax.loglog(x,y,'o')
ax.set_xlabel('Number of users per debate')
ax.set_ylabel('Normalized count')

fig.subplots_adjust(wspace=0.3)
```
